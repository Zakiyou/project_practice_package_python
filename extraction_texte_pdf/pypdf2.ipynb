```python
import PyPDF2

```

```python
# Fonction pour extraire le texte et les informations d'un fichier PDF
def extract_info_from_pdf(pdf_path):
    # Ouverture du fichier PDF en mode binaire
    with open(pdf_path, 'rb') as file:
        # Création d'un objet PDF reader
        pdf_reader = PyPDF2.PdfReader(file)
        
        # Nombre total de pages
        num_pages = len(pdf_reader.pages)
        
        # Extraction des métadonnées
        metadata = pdf_reader.metadata
        
        text = ""
        for page_num in range(num_pages):
            page = pdf_reader.pages[page_num]
            extracted_text = page.extract_text()
            if extracted_text:
                text += extracted_text
        
    return {
        'nombre_de_pages': num_pages,
        'metadonnees': metadata,
        'texte': text
    }
```

```python

pdf_path = "cv_developpeur.pdf"
info_pdf = extract_info_from_pdf(pdf_path)
```

```python

print("Informations sur le PDF :\n")

# Nombre de pages
print(f"Nombre de pages : {info_pdf['nombre_de_pages']}")

# Métadonnées
print("\nMétadonnées du PDF :")
if info_pdf['metadonnees']:
    for key, value in info_pdf['metadonnees'].items():
        # Les clés des métadonnées commencent souvent par '/'
        key_formate = key.lstrip('/')
        print(f"  {key_formate} : {value}")
else:
    print("  Aucune métadonnée trouvée.")

# Texte extrait
print("\nTexte extrait du PDF :\n")
print(info_pdf['texte'])

# Manipuler le texte, exemple : compter le nombre de mots
words = info_pdf['texte'].split()
print("\nNombre de mots dans le PDF :", len(words))
```
