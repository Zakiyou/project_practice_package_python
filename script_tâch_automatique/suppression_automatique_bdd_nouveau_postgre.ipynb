```python
import psycopg2
from psycopg2 import sql
```

```python

hostname = 'localhost'
username = 'postgres'
password = 'bagdadi'
port = '5432'
```

```python


try:
    conn = psycopg2.connect(
        dbname='postgres',  
        user=username,
        password=password,
        host=hostname,
        port=port
    )
    conn.autocommit = True  
    cur = conn.cursor()


    cur.execute("""
        SELECT datname 
        FROM pg_database 
        WHERE datname NOT IN ('postgres', 'template0', 'template1');
    """)

    databases = cur.fetchall()
    print(f"Bases de données à supprimer : {databases}")

    for database in databases:
        db_name = database[0]
        print(f"\nSuppression de la base de données : {db_name}")

        try:
            kill_conn_query = sql.SQL("""
                SELECT pg_terminate_backend(pg_stat_activity.pid)
                FROM pg_stat_activity
                WHERE pg_stat_activity.datname = %s
                AND pid <> pg_backend_pid();
            """)
            cur.execute(kill_conn_query, [db_name])
            print(f"Connexions actives sur {db_name} supprimées.")

            drop_db_query = sql.SQL("DROP DATABASE IF EXISTS {}").format(sql.Identifier(db_name))
            cur.execute(drop_db_query)
            print(f"Base de données {db_name} supprimée avec succès.")

        except Exception as e:
            print(f"Erreur lors de la suppression de {db_name}: {e}")

except Exception as main_e:
    print(f"Erreur lors de la connexion à PostgreSQL: {main_e}")

finally:
    try:
        if cur:
            cur.close()
    except:
        pass
    try:
        if conn:
            conn.close()
    except:
        pass
```
